```python
from flask import Flask, request, jsonify  
import pyodbc  
import threading 
import time
import pandas as pd
```

```python
app = Flask(__name__)
```

```python
data11 = (
    'DRIVER={ODBC Driver 17 for SQL Server};'
    'SERVER=DESKTOP-JGV3GCV\\SQLEXPRESS;'
    'DATABASE=RetailCustomer;'
    'Trusted_Connection=yes;'
)
```

```python
def get_db_connection():
    return pyodbc.connect(data11)
```

```python
@app.route('/A/upload_customers', methods=['POST'])
def upload_customers():
    if 'file' not in request.files:
        return jsonify({'error': 'No file uploaded'}), 400

    file = request.files['file']
    try:
        df = pd.read_csv(file)
        df.columns = df.columns.str.strip() 

        conn = get_db_connection()
        cursor = conn.cursor()

        for _, row in df.iterrows():
            cursor.execute("""
                INSERT INTO dbo.Customers(CustomerID, Name, Gender, Age, JoinDate, City)
                VALUES (?, ?, ?, ?, ?, ?)
            """, row['CustomerID'], row['Name'], row['Gender'], row['Age'], row['JoinDate'], row['City'])

        conn.commit()
        return jsonify({'message': f'{len(df)} customers inserted'}), 201

    except Exception as e:
        return jsonify({'error': str(e)}), 400
    finally:
        conn.close()
```
